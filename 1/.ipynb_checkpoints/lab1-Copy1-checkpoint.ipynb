```python
from itertools import product
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
imbook = cv2.imread('book.png')
imbook = cv2.cvtColor(imbook, cv2.COLOR_BGR2GRAY)
# print(imbook.shape)
```

```python
plt.imshow(imbook,cmap = "gray")
```

```python
from scipy import ndimage as flt
fltbook = flt.gaussian_filter(imbook, sigma = 30)
plt.imshow(fltbook,cmap = "gray")
```

```python
plt.imshow(imbook-fltbook, cmap='gray')
```

```python
import numpy as np
res = np.log10(imbook) - np.log10(fltbook)
print(np.array(res))
minres = np.min(res)
maxres = np.max(res)
res = (res-minres)*255/(maxres-minres)
res = res.astype(int)
print(minres,maxres)
plt.imshow(res, cmap='gray')
```

```python
def my_gaussian_filter(image, sigma):
    filt_razm = 2 * int(4 * sigma + 0.5) + 1  
    m = filt_razm//2
    n = filt_razm//2
    gaussian = np.zeros((filt_razm, filt_razm), np.float32)
    for x in range(-m, m+1):
        for y in range(-n, n+1):
            gaussian[x+m, y+n] = (1/(2*np.pi*(sigma**2)))*(np.exp(-(x**2 + y**2)/(2* sigma**2)))
    filt_iz = np.zeros_like(image, dtype=np.float32)
    for c in range(3):
        iznach_izob=image[:, :, 2]
        image_h = iznach_izob.shape[0]
        image_w = iznach_izob.shape[1]
        center_h = gaussian.shape[0]
        center_w = gaussian.shape[1]
        zapol_mass = np.pad(iznach_izob, pad_width=((center_h // 2, center_h // 2),(center_w // 2, center_w // 2)), mode='constant', constant_values=0).astype(np.float32)  
        h = center_h // 2
        w = center_w // 2
        preob_iz = np.zeros(zapol_mass.shape)
        for i in range(h, zapol_mass.shape[0]-h):
            for j in range(w, zapol_mass.shape[1]-w):
                
                x = zapol_mass[i-h:i-h+center_h, j-w:j-w+center_w]
                x = x.flatten()*gaussian.flatten()
                preob_iz[i][j] = x.sum()
        posl_h = -h
        posl_w = -w
        if(h == 0):
            return preob_iz[h:,w:posl_w]
        if(w == 0):
            return preob_iz[h:posl_h,w:]
        filt_iz[:, :, c] = preob_iz[h:posl_h,w:posl_w]
    return (filt_iz.astype(np.uint8))
```

```python
gaussian = my_gaussian_filter(imbook, sigma=30)
```

```python
plt.imshow(gaussian,cmap = "gray")
```

```python
plt.imshow(gaussian-imbook, cmap = "gray")
```

```python
import numpy as np
res = np.log10(imbook) - np.log10(gaussian)
minres = np.min(res)
maxres = np.max(res)
res = (res-minres)*255/(maxres-minres)
res = res.astype(int)
plt.imshow(res, cmap='gray')
```
